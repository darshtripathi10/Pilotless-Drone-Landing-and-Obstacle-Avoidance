```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
import torch

# Custom plot_one_box function
def plot_one_box(x, img, color=None, label=None, line_thickness=1, font_scale=0.5, font_thickness=1):
    # Coordinates
    c1, c2 = (int(x[0]), int(x[1])), (int(x[2]), int(x[3]))

    # Default color is blue if none provided
    color = color or [255, 0, 0]

    # Draw bounding box
    cv2.rectangle(img, c1, c2, color, thickness=line_thickness, lineType=cv2.LINE_AA)

    if label:
        # Get the label size
        t_size = cv2.getTextSize(label, 0, fontScale=font_scale, thickness=font_thickness)[0]

        # Coordinates for the label background box
        c2 = c1[0] + t_size[0], c1[1] - t_size[1] - 3

        # Draw label background box
        cv2.rectangle(img, c1, c2, color, -1, cv2.LINE_AA)

        # Put the label text
        cv2.putText(img, label, (c1[0], c1[1] - 2), 0, font_scale, [255, 255, 255], thickness=font_thickness, lineType=cv2.LINE_AA)

```

```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
import torch

# Custom plot_one_box function with smaller and clearer font
def plot_one_box(x, img, color=None, label=None, line_thickness=1, font_scale=0.5, font_thickness=1):
    # Coordinates
    c1, c2 = (int(x[0]), int(x[1])), (int(x[2]), int(x[3]))

    # Default color is blue if none provided
    color = color or [255, 0, 0]

    # Draw bounding box
    cv2.rectangle(img, c1, c2, color, thickness=line_thickness, lineType=cv2.LINE_AA)

    if label:
        # Get the label size
        t_size = cv2.getTextSize(label, cv2.FONT_HERSHEY_SIMPLEX, fontScale=font_scale, thickness=font_thickness)[0]

        # Coordinates for the label background box
        c2 = c1[0] + t_size[0], c1[1] - t_size[1] - 3

        # Draw label background box
        cv2.rectangle(img, c1, c2, color, -1, cv2.LINE_AA)

        # Put the label text
        cv2.putText(img, label, (c1[0], c1[1] - 2), cv2.FONT_HERSHEY_SIMPLEX, font_scale, [255, 255, 255], thickness=font_thickness, lineType=cv2.LINE_AA)

# Load custom YOLOv5 model
model = torch.hub.load('ultralytics/yolov5', 'custom', path='yolov5/runs/train/exp4/weights/last.pt', force_reload=True)

img = cv2.imread("/content/test2.jpg")

# Display the image
cv2_imshow(img)

img = cv2.imread("/content/test2.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Perform inference
results = model(img)


# Plot each bounding box with the custom function
for *box, conf, cls in results.xyxy[0]:  # xyxy format
    plot_one_box(box, img, label=f'{results.names[int(cls)]} {conf:.2f}', line_thickness=1, font_scale=0.3, font_thickness=1, color=(255, 0, 0))

# Display results
plt.figure(figsize=(10, 8))
plt.imshow(img)
plt.tight_layout()
plt.show()
```
